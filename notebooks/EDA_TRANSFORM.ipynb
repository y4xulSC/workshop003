# **Summary**

### World Happiness Datasets Comparison

The five World Happiness datasets share a similar structure but differ in data volume, column names, and null values. Below is a concise comparison.

---

### Similarities
- **Metrics**: Include Happiness Score, GDP per Capita, Freedom, Generosity, Social Support (or Family), and Perceptions of Corruption (or Trust).
- **Structure**: Each dataset has a country/region identifier, a happiness ranking, and numeric metrics with no duplicates and minimal null values.

---

### Differences

#### Number of Rows
- 2015: 158
- 2016: 157
- 2017: 157
- 2018, 2019: 156
- **Total**: 784

#### Number of Columns
- 2015: 12
- 2016: 13
- 2017: 12
- 2018, 2019: 9

#### Column Names
- 2015, 2016: Descriptive (e.g., `Economy (GDP per Capita)`).
- 2017: Dot notation (e.g., `Economy..GDP.per.Capita.`).
- 2018, 2019: Simplified (e.g., `GDP per capita`, `Social support`).

#### Additional Metrics
- 2015: Includes `Standard Error`.
- 2016: Includes `Lower` and `Upper Confidence Interval`.
- 2017: Includes `Whisker.high` and `Whisker.low`.
- 2018, 2019: Only core metrics, no intervals or errors.

---

### Variations
- **Major**: Column count and type differ; 2018 and 2019 have fewer variables; 2016 and 2017 include statistical metrics (confidence intervals, whiskers); 2015 includes standard error.
- **Minor**: Core metrics consistent across datasets (with varying names); data types are `float64` (metrics), `int64` (ranking), and `object` (country/region).

---

### Null Values
- Only 2018 has one null value in `Perceptions of corruption`.
- Other datasets are complete.

# **Dystopia Residual**

### Dropping the Dystopia Residual Column (2015,2016,2017)

**Reason**: Dystopia Residual is not an independent metric but an adjustment to align the sum of the six core variables (GDP per Capita, Social Support, Life Expectancy, Freedom, Generosity, Trust) with the Happiness Score.

For example, if the six metrics sum to 5.4 and the Happiness Score is 7.8, Dystopia Residual is 2.4 to bridge the gap. Since it is derived directly from the Score, it provides no new or independent information.

```python
cols_to_drop = ['Dystopia Residual', 'Dystopia.Residual']

for df in (df2015, df2016, df2017):
    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

df2015.head()
```

# Region:

The 2017, 2018, and 2019 datasets lack a region column. We will add it using the country-region mappings from the 2015 and 2016 datasets.

2015 with region

```python
df2015.head()
```

2019 without region

```python
df2019.head()
```

Add region

```python
import pandas as pd

region_map = (
    df2015.set_index('country')['region']
    .combine_first(df2016.set_index('country')['region'])
)

for df in (df2017, df2018, df2019):
    df['region'] = df['country'].map(region_map)

cols = [
    'country','region','score','gdp_per_capita',
    'social_support','health_life_expectancy',
    'freedom','trust_government_corruption','generosity'
]

df2015 = df2015[cols]
df2016 = df2016[cols]
df2017 = df2017[cols]
df2018 = df2018[cols]
df2019 = df2019[cols]
```

**Null Values**: The 2018 and 2019 datasets have null values for regions, but we will manually assign them based on the regions listed in the 2015 dataset.

```python
manual_regions = {
    'Trinidad & Tobago': 'Latin America and Caribbean',
    'Northern Cyprus': 'Middle East and Northern Africa',
    'North Macedonia': 'Central and Eastern Europe',
    'Gambia': 'Sub-Saharan Africa'
}

for df in (df2018, df2019):
    df['region'] = df['region'].fillna(df['country'].map(manual_regions))
```

# **Merge**

```python
merge = pd.concat(
    [df2015, df2016, df2017, df2018, df2019],
    ignore_index=True
)
```

```python
merge['trust_government_corruption'] = merge.groupby('country')['trust_government_corruption'] \
    .transform(lambda x: x.fillna(x.median()))

merge = merge.drop_duplicates(subset=['country','year'], keep='first').reset_index(drop=True)
```
